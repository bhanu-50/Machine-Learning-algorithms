```
def line_plot(w,w0,X,Y):
  plt.scatter(X,Y)
  plt.plot(X,w*X+w0)
  plt.show()
```

```
line_plot(2.4,4,X,Y)
```

```
#Loss function
def loss_function(w,w0,X,Y):
  y_hat = np.dot(w,X)+w0
  return np.sum((Y-(y_hat))**2)
```

```
def gradient(w,w0,X,Y):
  y_hat = np.dot(w,X)+w0
  return -np.sum(2*X*(Y-y_hat))/len(X),-np.sum(2*(Y-y_hat))/len(X)
```

```
def gradient_descent(X,Y,w,w0,lr = 0.01,ITR = 1000):
  for i in range(ITR):
    gr,gr0 = gradient(w,w0,X,Y)
    w = w - lr*gr
    w0 = w0 - lr*gr0
    print(w,w0)
    if abs(gr)<0.01 and abs(gr0)<0.01:
      line_plot(w,w0,X,Y)
      print(f'{w},{w0}')
      break
```

```
gradient_descent(X,Y,100,200,lr = 0.01,ITR = 1000)
```
