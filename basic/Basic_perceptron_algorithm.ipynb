```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
from sklearn import datasets
```

```
X,Y = datasets.make_blobs(n_samples = 200,n_features = 2,centers = 2,center_box=(0,10))
```

```
X,Y
```

```
Y[Y==0]=-1
```

```
plt.plot(X[:,0][Y==-1],X[:,1][Y==-1],'g^')
plt.plot(X[:,0][Y==1],X[:,1][Y==1],'bs')
plt.show()
```

```
w = np.random.normal(size=(2,))
```

```
w
```

```
w0 = np.random.normal(size=(1,))
```

```
w0
```

```
def display(w,w0,X,Y):
  intercept = -w0/w[1]
  slope = -w[0]/w[1]
  line_x = np.arange(-10,10)
  line_y = slope*line_x + intercept
  plt.plot(X[:,0][Y==-1],X[:,1][Y==-1],'g^')
  plt.plot(X[:,0][Y==1],X[:,1][Y==1],'bs')
  plt.plot(line_x,line_y,'r-')
  plt.xlabel("X1")
  plt.ylabel("X2")
  plt.show()
```

```
display(w,w0,X,Y)
```

```
def persptron(X,Y,w,w0,lr=0.1,):
  n=len(X)
  for it in range(10001):
    mis_clf = 0
    for i in range(n):
      half_space = np.dot(w,X[i])+w0
      if half_space*Y[i]<0:
        mis_clf += 1
        w = w + lr * X[i]*Y[i]
        w0 = w0 + lr * Y[i]
    acc = (n-mis_clf)/n
    if it %1000==0:
      display(w,w0,X,Y)
      print(f"Accuracy:{acc}")
  return w,w0

```

```
persptron(X,Y,w,w0)
```
